```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# machine learning imports
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score

# preprocessing imports
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
X = df[['Layout', 'Powder', 'Location', 'PlateID']]
X
```

```python
y = df['Nonconformity']
y
```

```python
# splitting the train data into train and validation sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# preprocessing the data (scaling, imputing, encoding)
categorical_features = ['Layout', 'Powder', 'Location', 'PlateID']

cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", dtype=np.int8)
)

preprocessor = make_column_transformer(
    (cat_pipe, categorical_features),
    remainder="drop"
)
```

```python
Cs = np.logspace(-3, 2, 12)  # 0.001 → 100
rows = []

for C in Cs:
    pipe = Pipeline(steps=[
        ("pre", preprocessor),
        ("clf", LogisticRegression(
            penalty="l1",
            solver="saga",        # must use saga for L1
            C=C,
            #max_iter=5000,
            class_weight="balanced"  # helpful if classes are imbalanced
        ))
    ])
    pipe.fit(X_train, y_train)
    y_prob = pipe.predict_proba(X_test)[:, 1]
    y_pred = (y_prob >= 0.5).astype(int)

    rows.append({
        "C": C,
        "roc_auc": roc_auc_score(y_test, y_prob),
        "f1": f1_score(y_test, y_pred),
        "accuracy": accuracy_score(y_test, y_pred),
        # sparsity: how many nonzero coefficients after one-hot
        "nnz": (pipe.named_steps["clf"].coef_ != 0).sum()
    })

results = pd.DataFrame(rows).sort_values(["roc_auc","f1"], ascending=False)
print(results)

# best C (by ROC AUC here)
best_C = results.iloc[0]["C"]
print("Best C by AUC:", best_C)
```

```python
best_pipe = Pipeline(steps=[
    ("pre", preprocessor),
    ("clf", LogisticRegression(
        penalty="l1", solver="saga", C=best_C,
        class_weight="balanced", max_iter=5000
    ))
])
best_pipe.fit(X_train, y_train)

# 2) Get per-row probabilities on the TEST set (holdout)
y_prob_test = best_pipe.predict_proba(X_test)[:, 1]
y_prob_test

res = X_test.copy()
res["Scrap_Prob"] = y_prob_test
res["Actual"] = y_test.values 
res[res['Actual']==True]
results[['C','nnz']]
```

```python
# 1) Refit at best_C
best_pipe = Pipeline(steps=[
    ("pre", preprocessor),
    ("clf", LogisticRegression(
        penalty="l1", solver="saga",
        C=float(best_C),
        max_iter=20000, tol=1e-3,
        class_weight="balanced",
        random_state=42
    ))
])
best_pipe.fit(X_train, y_train)

# 2) Get feature names after preprocessing (from the FITTED ColumnTransformer)
ct = best_pipe.named_steps["pre"]               # the fitted ColumnTransformer
feature_names = ct.get_feature_names_out()      # works across transformers (incl. OHE)

# 3) Get coefficients from the fitted logistic regression
coefs = best_pipe.named_steps["clf"].coef_.ravel()

# 4) Build coef table and show strongest signals
import pandas as pd
coef_df = pd.DataFrame({"feature": feature_names, "coef": coefs})
coef_df_nonzero = coef_df[coef_df["coef"] != 0]

print("Top + predictors (push toward Nonconformity=1):")
print(coef_df_nonzero.sort_values("coef", ascending=False).head(10))

print("\nTop − predictors (push toward Nonconformity=0):")
print(coef_df_nonzero.sort_values("coef", ascending=True).head(10))
```
